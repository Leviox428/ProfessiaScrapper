## threadpoolexecutor

```python
import time
from concurrent.futures import ThreadPoolExecutor

def simulate_work(duration, task_id):
    print(f"Task {task_id} starts.")
    time.sleep(duration)
    print(f"Task {task_id} completed.")
    return f"Task {task_id} result."

def run_sequential():
    durations = [2, 3, 1]  # Example task durations
    results = []
    start_time = time.time()
    for i, duration in enumerate(durations, start=1):
        results.append(simulate_work(duration, i))
    end_time = time.time()
    print("Sequential execution time:", end_time - start_time)
    return results

# Multithreaded execution using ThreadPoolExecutor
def run_multithreaded():
    durations = [2, 3, 1]  # Example task durations
    task_ids = range(1, len(durations) + 1)
    start_time = time.time()
    with ThreadPoolExecutor(max_workers=3) as executor:
        # Use executor.map to run tasks in parallel
        results = list(executor.map(simulate_work, durations, task_ids))
    end_time = time.time()
    print("Multithreaded execution time:", end_time - start_time)
    return results

print("\nRunning tasks sequentially:")
sequential_results = run_sequential()

print("\nRunning tasks with multithreading:")
multithreaded_results = run_multithreaded()
```
